```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"Downloads\Sample - Superstore.csv", encoding = 'ISO-8859-1')
df
```

```python
df1= df.dropna()
df1
```

```python
null_check=df1.isna().sum()
null_check
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figure=(10,6))
```

```python
sns.scatterplot(x='Sales', y='Profit', data=df1)
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figure=(10,6))
```

```python
sales_over_time = df.groupby('Order Date')['Sales'].sum().reset_index()
sns.lineplot(x='Order Date', y='Sales', data= sales_over_time)
plt.title("Orderdate vs Sales")
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figure=(10,6))
sns.barplot(x='Category', y='Sales', hue='Region', data=df1)
plt.title("category vs region")
plt.xlabel("category")
plt.ylabel("region")
plt.show()
```

```python
sns.scatterplot(x='Quantity', y='Profit', data=df1)
plt.title("Quantity vs Profit")
plt.xlabel("Quantity")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figure=(10,6))
sns.barplot(y='Sales', x='Ship Mode',data=df1)
plt.ylabel('Sales')
plt.xlabel('Quantity')
plt.show()
```

```python
import pandas as pd
mean1=df1['Sales'].mean()
median1=df1['Sales'].median()
mode1=df1['Sales'].mode()

print("Mean of Sales:",mean1)
print("Median of Sales:",median1)
print("Mode of Sales:",mode1)
```

```python
import pandas as pd
mean2=df1['Profit'].mean()
median2=df1['Profit'].median()
mode2=df1['Profit'].mode()

print("Mean of profit:",mean2)
print("Median of profit:",median2)
print("Mode of profit:",mode2)
```

```python
import pandas as pd
mean1=df1['Discount'].mean()
median1=df1['Discount'].median()
mode1=df1['Discount'].mode()

print("Mean of Discount:",mean1)
print("Median of Discount:",median1)
print("Mode of Discount:",mode1)
```
